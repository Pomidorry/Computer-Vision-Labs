```python
import dlib   
import cv2   
import matplotlib.pyplot as plt  
import numpy as np
```

```python
image_path = 'C:\labsCV\lab6\scary.jpg'
img = cv2.imread(image_path)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
detector = dlib.get_frontal_face_detector()
```

```python
faces = detector(img_rgb)
print(f"Number of faces detected: {len(faces)}")
```

```python
def rect_to_bb(rect):
    # Dlib rect --> OpenCV rect
    x = rect.left()
    y = rect.top()
    w = rect.right() - x
    h = rect.bottom() - y

    return (x, y, w, h)
```

```python
result_dlib = np.copy(img)
faces_dlib_img = []
for rect in faces:    
    x, y, w, h = rect_to_bb(rect)
    print(x, y, w, h)
    cv2.rectangle(result_dlib, (x, y), (x+w, y+h), (0, 255, 0), 3)
    faces_dlib_img.append(img[y:y+h, x:x+w, :])
    

plt.subplot(121), plt.imshow(img), plt.title('Original')
plt.subplot(122), plt.imshow(result_dlib), plt.title('dlib')
```
